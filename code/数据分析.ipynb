```python
plt.figure(figsize = (20,16))
colnm = train_data.columns.tolist()[:39]
mcorr = train_data[colnm].corr()
mask = np.zeros_like(mcorr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220, 10, as_cmap=True)
g = sns.heatmap(mcorr, mask=mask, cmap=cmap, square=True, annot=True, fmt='0.2f')
```

2. 有些变量之间有很强的相关性。

```python
# 看一下lightgbm特征重要性排名
import lightgbm as lgb
```

```python
model_lgb = lgb.LGBMRegressor(random_state=2018)
feature_col = [v for v in train_data.columns if v not in ['target']]
X_train = train_data[feature_col].values
y_train = train_data.target.values
```

```python
model_lgb.fit(X_train, y_train)
```

```python
print(pd.DataFrame({'column': feature_col, 'importance': model_lgb.feature_importances_}).sort_values(by='importance', ascending=False))
```
